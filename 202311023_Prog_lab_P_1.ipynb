```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/january.csv')
```

```python
def date_col(df,master_df):
  x=df.shape[1]
  new=[]
  for i in range(1,x):
    for j in range(96):
      new.append(i)

  master_df['date'] = new
```

```python
def fill(df,master_df):
  def day_or_night(hour):
    if 6 <= hour <= 18:
        return 'day'
    else:
        return 'night'

  master_df[['Hour', 'Minute']] = df['Time'].str.split(':', expand=True)

  # Convert 'Hour' and 'Minute' columns to integers
  master_df['Hour'] = master_df['Hour'].astype(int)
  master_df['Minute'] = master_df['Minute'].astype(int)

  # Apply the function to create a new 'Day_Night' column
  master_df['time_of_day'] = master_df['Hour'].apply(day_or_night)
  master_df['year'] = master_df['year'].fillna(2022)
  master_df['month'] = master_df['month'].fillna(12)

```

```python
def master(df):
  new_df = df.drop('Time',axis=1)
  master_df = pd.DataFrame(columns=['frequency','date','month','year','Hour','Minute','time_of_day'])
  master_df['frequency'] = pd.concat([new_df[col] for col in new_df.columns], axis=0)
  date_col(df,master_df)
  fill(df,master_df)
  return master_df
```

```python
# this function helps converting the data from df to master_df
master(df)

# master_df
```
